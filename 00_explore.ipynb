```python
# Notebook: Análise Exploratória das Variáveis Climáticas (tipo e diversidade)

import pandas as pd

# === Carregar os dados de clima === #
weather = pd.read_csv("data/weather/weather.csv")

# Converter datetime se ainda não estiver
weather['datetime'] = pd.to_datetime(weather['datetime'])

# Selecionar apenas colunas úteis
cols_to_check = ['preciptype', 'conditions', 'icon', 'stations']

# Ver tipos únicos de cada coluna categórica
for col in cols_to_check:
    print(f"\n\033[1mValores únicos em {col} (total={weather[col].nunique()}):\033[0m")
    print(weather[col].dropna().unique())

# Analisar possíveis valores extremos e tipos de variáveis contínuas
print("\n\033[1mResumo Estatístico de Variáveis Contínuas:\033[0m")
print(weather.describe().T[['min', 'max']])
```
